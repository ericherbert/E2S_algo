 # <font color=#df062a> <u>I. Introduction</u></font>
   ## <font color=#2d06df><i>I. 1) Qu'est-ce qu'une fonction python?</i></font>
<div class="alert alert-success" style="border-left: 5px solid #3c763d;border-radius: 5px;">
    Nous avons jusqu'ici écrit et exécuté des programmes que la machine <strong>exécutait ligne par ligne</strong>.<br>
  Certains programmes peuvent être très longs. On va chercher à écrire nos programmes de la manière la plus <strong>compacte possible</strong>, en utilisant des <strong>fonctions</strong>. <br>
 
 <strong><u>Définition:</u></strong><ul>
     <li> Une <strong>fonction</strong> en Python est un petit programme permettant de représenter l'action d'une <strong>application</strong> qui, étant donnés des <strong>arguments</strong>, renvoie un <strong>résultat</strong>.
         <li>
<li>L'usage de fonctions permet de diviser un programme en plusieurs <strong>tâches intermédiaires</strong>.
    </ul>
    
Il est possible de créer soi-même des fonctions via la syntaxe suivante :<br>
<code>def nom_de_la_fonction (nom_des_paramètres) :
instructions
return résultat</code>

</div>
<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Vous connaissez déjà une fonction : la fonction <code>print</code>, dont l'argument est une <strong>chaîne de caractères</strong>, et dont l'action est d'afficher une phrase.
    </div>

<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Le petit programme ci-dessous code la fonction <code>triple</code> qui, étant donné un nombre réel $x$, renvoie le triple de ce nombre, $3x$.
    </div>

```python
# Pour définir la fonction, on écrit "def" pour définition, 
# suivi du nom de la fonction, ici "triple", et le ou les arguments entre parenthèses, ici juste x
def triple(x):
    # On effectue l'opération 3 fois x, que l'on doit stocker quelque part! 
    y = 3*x
    # On "renvoie" (return) le résultat du calcul
    return y
```

<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Testons cette fonction! On peut stocker le résultat obtenu dans une variable puis l'afficher. Exécutons la cellule suivante.
    </div>

```python
x = 5
a = triple(x)
print("Le triple de ",x," vaut ",a,"\n")
```

<div class="alert alert-warning"><strong>Exemple :</strong><br>
     On peut aussi directement appeler la fonction. Exécutons la cellule suivante.
    </div>

```python
print("Le triple de 10 vaut ",triple(10),"\n")
```

<div class="alert alert-warning"><strong>Exemple :</strong><br>
   Et aussi utiliser la fonction dans des calculs! Exécutons la cellule suivante.
    </div>

```python
x=2
b = 7+triple(9)*x;
print(b)
```

```python
def maximum(a, b, c):
    m = a
    if b > m:
        m = b
    if c > m:
        m = c
    return m

result = maximum(3 * math.pi, math.sqrt(85), 5325.47)
print("Le maximum entre 3𝜋, 85√ et 5325.47 est :", result)
```

<div class="alert alert-info"><strong>Exercice 7 </strong><br> La grand-mère de Solal lui a donné 50 € pour son anniversaire qu'il met dans sa tirelire. Ses parents lui donnent 15 € d'argent de poche chaque mois. Solal décide d'économiser jusqu'à ce qu'il puisse acheter la guitare de ses rêves qui coûte 490 €. <br>
Compléter le script de la fonction <code>guitare</code> ci-dessous afin qu'elle renvoie le nombre de mois que devra attendre Solal avant de s'acheter sa guitare.

</div>

```python
def guitare():
    m = 0
    tirelire = 50
    while tirelire < 490:
        tirelire += 15
        m += 1
    return m

mois_attente = guitare()
print("Solal devra attendre", mois_attente, "mois avant de s'acheter sa guitare.")
```

<div class="alert alert-info"><strong>Exercice 8 </strong>  <br> <ol>
<li>Dans la cellule suivante, ajouter une ligne de code pour appeler la fonction mystère avec des entiers $p$ et $q$ non nuls arbitraires.
<li> Que fait cette fonction ?
</ol>
</div>

```python
def fonction_mystere(p, q):
    print("Les entiers choisis sont p=", p, "et q=", q)
    div = 1
    if p > q:
        m = p
    else:
        m = q
    for i in range(2, m + 1):
        if p % i == 0 and q % i == 0:
            div = i
    print(p, "/", q)
    p = p / div
    q = q / div
    print("=", p, "/", q)
    return "Opération terminée"

print(fonction_mystere(7, 3))
```

<div class="alert alert-info"><strong>Exercice 9  </strong><br> Un client place le 1er janvier 2022 une somme de 2 000 € sur un compte bancaire. Le compte rapporte de l'argent chaque année en appliquant un taux de 2 %. Le client attend que la somme d'argent sur le compte dépasse 2 700 € pour retirer la totalité de l'argent. On considère le script suivant :
</div>

```python
def compte():
    somme=2000
    n=0
    while somme<=2700:
        somme=1.02*somme
        n=n+1
    return 2022+n
```

<div class="alert alert-warning"><strong>Remarque :</strong><br>
    <ol>
        <li>La variable <code>n</code> est un compteur : elle est initialisée à 0 et permet de compter le nombre de fois où l'on va effectuer l'augmentation.
        <li>La fonction <code>compte</code> n'a pas d'argument.
    </ol>
    </div>

<div class="alert alert-info">Exécuter l'instruction ci-dessous et interpréter le résultat renvoyé par la fonction <code>compte</code>.
</div>

```python
compte()
```

<strong>Réponse :</strong> Il devra attendre l'année 2038pour etre payé 2700 euros

<div class="alert alert-info">On modifie maintenant la fonction <code>compte</code> avec un argument:</div>

```python
def compte(k):
    somme=2000
    n=0
    while somme<=k:
        somme=1.02*somme
        n=n+1
    return 2022+n
compte(1500)
```
