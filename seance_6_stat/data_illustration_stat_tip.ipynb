```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# reading the database
data = pd.read_csv( "tips.csv" )
 
# printing the top 10 rows
display(data.head(10))

```

```python

# Scatter plot with day against tip
figname = 'test correlation'
plt.figure(figname)
plt.subplot(2,2,1)
plt.scatter( data['total_bill'] , data['tip'] )
plt.subplot(2,2,2)
plt.scatter( data['day'] , data['tip'] )
plt.subplot(2,2,3)
plt.scatter( data['sex'] , data['tip'] )
plt.subplot(2,2,4)
plt.scatter( data['size'] , data['total_bill'] )

# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
#plt.xlabel('Day')
#plt.ylabel('Tip')

plt.tight_layout()

plt.show()
```

```python
 
# Scatter plot with day against tip
plt.scatter(data['day'], 
            data['tip'], 
            c = data['size'], 
            s = data['total_bill'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
plt.colorbar()
 
plt.show()

```

```python

# Scatter plot with X vs Y
plt.plot(data['smoker'], data['size'], 'o')
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('Size')
plt.ylabel('total bill')
 
plt.show()
```

```python

 
# Bar chart with day against tip
plt.bar(data['day'], data['tip'])
 
plt.title("Bar Chart")
 
# Setting the X and Y labels
plt.xlabel('Day')
plt.ylabel('Tip')
 
# Adding the legends
plt.show()
```

```python

data = pd.read_csv( "tips.csv" )

# histogram of total_bills
var = 'total_bill'
data[var] = data[var] / data['size']
plt.hist( data[var], bins=10, density=0)
 
plt.title( var )
plt.ylabel('f')
plt.xlabel(var)
# Adding the legends

std = np.std(data[var])
mean = np.mean(data[var])
print('Ecart type : ' + str(std))
print('Moyenne : ' + str(mean))

x = np.linspace(min(data[var]), max(data[var]), 100)
import scipy as sc
y = sc.stats.norm.pdf( x, mean, std)

plt.plot(x,y,'r')

plt.show()
```
