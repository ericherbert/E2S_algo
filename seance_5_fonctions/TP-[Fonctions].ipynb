 # <font color=#df062a> <u>I. Introduction</u></font>
   ## <font color=#2d06df><i>I. 1) Qu'est-ce qu'une fonction python?</i></font>
<div class="alert alert-success" style="border-left: 5px solid #3c763d;border-radius: 5px;">
    Nous avons jusqu'ici √©crit et ex√©cut√© des programmes que la machine <strong>ex√©cutait ligne par ligne</strong>.<br>
  Certains programmes peuvent √™tre tr√®s longs. On va chercher √† √©crire nos programmes de la mani√®re la plus <strong>compacte possible</strong>, en utilisant des <strong>fonctions</strong>. <br>
 
 <strong><u>D√©finition:</u></strong><ul>
     <li> Une <strong>fonction</strong> en Python est un petit programme permettant de repr√©senter l'action d'une <strong>application</strong> qui, √©tant donn√©s des <strong>arguments</strong>, renvoie un <strong>r√©sultat</strong>.
         <li>
<li>L'usage de fonctions permet de diviser un programme en plusieurs <strong>t√¢ches interm√©diaires</strong>.
    </ul>
    
Il est possible de cr√©er soi-m√™me des fonctions via la syntaxe suivante :<br>
<code>def nom_de_la_fonction (nom_des_param√®tres) :
instructions
return r√©sultat</code>

</div>
<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Vous connaissez d√©j√† une fonction : la fonction <code>print</code>, dont l'argument est une <strong>cha√Æne de caract√®res</strong>, et dont l'action est d'afficher une phrase.
    </div>

<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Le petit programme ci-dessous code la fonction <code>triple</code> qui, √©tant donn√© un nombre r√©el $x$, renvoie le triple de ce nombre, $3x$.
    </div>

```python
# Pour d√©finir la fonction, on √©crit "def" pour d√©finition, 
# suivi du nom de la fonction, ici "triple", et le ou les arguments entre parenth√®ses, ici juste x
def triple(x):
    # On effectue l'op√©ration 3 fois x, que l'on doit stocker quelque part! 
    y = 3*x
    # On "renvoie" (return) le r√©sultat du calcul
    return y
```

<div class="alert alert-warning"><strong>Exemple :</strong><br>
    Testons cette fonction! On peut stocker le r√©sultat obtenu dans une variable puis l'afficher. Ex√©cutons la cellule suivante.
    </div>

```python
x = 5
a = triple(x)
print("Le triple de ",x," vaut ",a,"\n")
```

<div class="alert alert-warning"><strong>Exemple :</strong><br>
     On peut aussi directement appeler la fonction. Ex√©cutons la cellule suivante.
    </div>

```python
print("Le triple de 10 vaut ",triple(10),"\n")
```

<div class="alert alert-info"><strong>Exercice 7 </strong><br> La grand-m√®re de Solal lui a donn√© 50 ‚Ç¨ pour son anniversaire qu'il met dans sa tirelire. Ses parents lui donnent 15 ‚Ç¨ d'argent de poche chaque mois. Solal d√©cide d'√©conomiser jusqu'√† ce qu'il puisse acheter la guitare de ses r√™ves qui co√ªte 490 ‚Ç¨. <br>
Compl√©ter le script de la fonction <code>guitare</code> ci-dessous afin qu'elle renvoie le nombre de mois que devra attendre Solal avant de s'acheter sa guitare.

</div>

```python
def guitare():
    m = 0
    tirelire = 50
    while tirelire < 490:
        tirelire += 15
        m += 1
    return m

mois_attente = guitare()
print("Solal devra attendre", mois_attente, "mois avant de s'acheter sa guitare.")
```

<div class="alert alert-info"><strong>Exercice 9  </strong><br> Un client place le 1er janvier 2022 une somme de 2 000 ‚Ç¨ sur un compte bancaire. Le compte rapporte de l'argent chaque ann√©e en appliquant un taux de 2 %. Le client attend que la somme d'argent sur le compte d√©passe 2 700 ‚Ç¨ pour retirer la totalit√© de l'argent. On consid√®re le script suivant :
</div>

```python
def compte():
    somme=2000
    n=0
    while somme<=2700:
        somme=1.02*somme
        n=n+1
    return 2022+n
```

    <ol>
